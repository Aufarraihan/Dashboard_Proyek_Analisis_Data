## Visualization & Explanatory Analysis

### Bagaimana Performa Penjualan dan Revenue Perusahaan dalam Beberapa Bulan Terakhir?

```python
monthly_orders_df = all_df.resample(rule='m', on="order_approved_at_y").agg({
    "order_id": "nunique",
    "price_y": "sum"
})

monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "price_y": "revenue"
}, inplace=True)
monthly_orders_df
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(monthly_orders_df["order_approved_at_y"], monthly_orders_df["order_count"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Number of Orders per Month (2018)", loc="center", fontsize=20) 
for x, y in zip(monthly_orders_df["order_approved_at_y"], monthly_orders_df["order_count"]):
    plt.text(x, y, str(y), fontsize=10, ha='center', va='bottom')
plt.xticks(fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_orders_df["order_approved_at_y"],
    monthly_orders_df["revenue"],
    marker='o', 
    linewidth=2,
    color="#72BCD4"
)
for x, y in zip(monthly_orders_df["order_approved_at_y"], monthly_orders_df["revenue"]):
    plt.text(x, y, f"{y:,.0f}", fontsize=10, ha='center', va='bottom')
plt.title("Total Revenue per Month (2018)", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### Apa produk kategori yang menghasilkan revenue paling banyak dan paling sedikit dalam beberapa bulan trakhir?


```python
sum_revenue_category_df = all_df.groupby("product_category_name_english").agg({
    "order_id": "nunique",
    "price_y": "sum"
}).sort_values(by="price_y", ascending=False).reset_index()
sum_revenue_category_df.rename(columns={
    "order_id": "order_count",
    "price_y": "revenue"
}, inplace=True)
sum_revenue_category_df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Best Performing Product
best_plot = sns.barplot(x="revenue", y="product_category_name_english", data=sum_revenue_category_df.head(6), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing Category Product", loc="center", fontsize=15)
ax[0].tick_params(axis='y', labelsize=12)

# Menambahkan label angka di atas bar
for container in best_plot.containers:
    best_plot.bar_label(container, fmt="%d", fontsize=12, padding=3)

# Worst Performing Product
worst_plot = sns.barplot(x="revenue", y="product_category_name_english", data=sum_revenue_category_df.sort_values(by="revenue", ascending=True).head(6), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

# Menambahkan label angka di atas bar
for container in worst_plot.containers:
    worst_plot.bar_label(container, fmt="%d", fontsize=12, padding=3)

plt.suptitle("Best and Worst Performing Category Product by Revenue", fontsize=20)
plt.show()
```

### Bagaimana revenue dari pelanggan yang kita miliki disetiap negara dalam beberapa bulan terakhir?


```python
bystate_df = all_df.groupby(by="customer_state").agg({
    "customer_id_x": "nunique",
    "price_y": "sum"
}).sort_values(by="price_y", ascending=False).reset_index()
bystate_df.rename(columns={
    "customer_id_x": "order_count",
    "price_y": "revenue"
}, inplace=True)
bystate_df
```

```python
plt.figure(figsize=(10, 5))

colors_ = ["#72BCD4"] + ["#D3D3D3"] * 19  

barplot = sns.barplot(
    x="revenue", 
    y="customer_state",
    data=bystate_df.sort_values(by="revenue", ascending=False).head(20),
    palette=colors_
)

for container in barplot.containers:
    barplot.bar_label(container, fmt="%d", fontsize=10, padding=3)


plt.title("Number of Customers by State", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

### Kapan terakhir pelanggan melakukan transaksi?

```python
rfm_df = all_df.groupby(by="customer_id_x", as_index=False).agg({
    "order_approved_at_y": "max", # mengambil tanggal order terakhir
    "order_id": "nunique", # menghitung jumlah order
    "price_y": "sum" # menghitung jumlah revenue yang dihasilkan
})
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]
 
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = orders_df["order_approved_at"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)
 
rfm_df.sort_values(by="max_order_timestamp", ascending=False).head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

colors = ["#72BCD4"] * 5

sns.barplot(y="recency", x="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax)
ax.set_ylabel(None)
ax.set_xlabel(None)
ax.set_title("By Recency (days)", loc="center", fontsize=18)
ax.tick_params(axis='x', labelsize=15)
ax.set_xticks(ax.get_xticks()) 
ax.set_xticklabels(ax.get_xticklabels(), rotation=315, ha="left")

plt.tight_layout()
plt.subplots_adjust(bottom=0.2)
plt.show()
```

### Seberapa sering seorang pelanggan melakukan pembelian dalam beberapa bulan terakhir?

```python
fig, ax = plt.subplots(figsize=(10, 6))

colors = ["#72BCD4"] * 5

sns.barplot(y="frequency", x="customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(), palette=colors, ax=ax)
ax.set_ylabel(None)
ax.set_xlabel(None)
ax.set_title("By Frequency", loc="center", fontsize=18)
ax.tick_params(axis='x', labelsize=15, rotation=45)
ax.set_xticks(ax.get_xticks()) 
ax.set_xticklabels(ax.get_xticklabels(), rotation=315, ha="left")

plt.tight_layout()
plt.subplots_adjust(bottom=0.2)
plt.show()
```

### Berapa banyak uang yang dihabiskan pelanggan dalam beberapa bulan terakhir? 

```python
fig, ax = plt.subplots(figsize=(10, 6))

colors = ["#72BCD4"] * 5

barplot = sns.barplot(y="monetary", x="customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax)
for container in barplot.containers:
    barplot.bar_label(container, fmt="%.0f", fontsize=12, padding=3)
ax.set_ylabel(None)
ax.set_xlabel(None)
ax.set_title("By Monetary", loc="center", fontsize=18)
ax.tick_params(axis='x', labelsize=15, rotation=45)
ax.set_xticks(ax.get_xticks()) 
ax.set_xticklabels(ax.get_xticklabels(), rotation=315, ha="left")

plt.tight_layout()
plt.subplots_adjust(bottom=0.2)
plt.show()
```
